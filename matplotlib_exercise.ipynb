## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv('data/hist_data.csv')

#bins(棒の数) alpha(透明度) color(色) label(凡例に表示されるラベル)
plt.hist(df_hist['aデータ'],bins = 30, alpha = 0.5, color = 'orange', label = 'date a')
plt.hist(df_hist['bデータ'],bins = 30, alpha = 0.5, color = 'c', label = 'data b')

#Frequency(度数) tempreture(温度)
plt.title('Frequency distribution regarding temperature of a data and b data')
plt.xlabel('temperature(℃)')
plt.ylabel('frequency')

#凡例表示
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv('data/scatter_data.csv')
df_class0 = df_scatter[df_scatter['class'] == 0]
df_class1 = df_scatter[df_scatter['class'] == 1]
df_class2 = df_scatter[df_scatter['class'] == 2]
label = df_scatter.columns

plt.scatter(df_class0['length (cm)'], df_class0['width (cm)'], color = 'red', label = 'class0')
plt.scatter(df_class1['length (cm)'], df_class1['width (cm)'], color = 'blue', label = 'class1')
plt.scatter(df_class2['length (cm)'], df_class2['width (cm)'], color = 'green', label = 'class2')

plt.title('Scatter plot for length and width of classes 0~2')
plt.xlabel(label[0])
plt.ylabel(label[1])
plt.grid()
plt.legend()

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv('data/line_data.csv')
df_line["期間"] = ['2016 Second half','2017 First half','2017 Second half','2018 First half',
                 '2018 Second half','2019 First half','2019 Second half','2020 First half']


plt.plot(df_line['期間'],df_line['件数'], marker = 'o')
plt.xticks(df_line['期間'], rotation=30)

plt.title('Changes in the number of cases by period')
plt.xlabel('period')
plt.ylabel('number')
plt.grid()
plt.show()

plt.bar(df_line['期間'],df_line['費用'])
plt.xticks(df_line['期間'], rotation=30)


plt.title('Changes in the cost of cases by period')
plt.xlabel('period')
plt.ylabel('cost(yen)')
plt.grid()

plt.show()
```
